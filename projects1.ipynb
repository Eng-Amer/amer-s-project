```python
# calculator
w = 0

def startScreen():
  print("\n#######################################################")
  print("#                   welcome sir                         #")
  print("#                enter your process                     #")
  print("#                                                       #")
  print("   1-addition                          4-multiplication #")
  print("#  2-division                          5-modulus        #")
  print("#  3-substraction                      6-factorial      #")
  od= int(input(                 ))
  print("#                                                       #")
  print("#########################################################")
  if od == 1:
    addition()
  elif od == 2:
    division()
  elif od == 3:
    substaction()
  elif od == 4:
    multiplication()
  elif od == 5:
    modulus()
  elif od == 6:
    factorial(w)
  else:
    errorScreen()
  return  
  







def addition():
  x = float(input("enter your first number"))
  d = float(input("enter your second number"))
  print(x+d)
  startScreen()
  return


def division():
  x = float(input("enter your first number"))
  d = float(input("enter your second number"))
  if d != 0:
    print(x/d)
    startScreen()
  elif x != 0:
    print(d/x)
    startScreen()
  else:
    errorScreen()

  return


def substaction():
  x = float(input("enter your first number"))
  d = float(input("enter your second number"))
  print(x-d)
  startScreen()
  return

def multiplication():
  x = float(input("enter your first number"))
  d = float(input("enter your second number"))
  print(x*d)
  startScreen()
  return

def modulus():
  x = float(input("enter your first number"))
  d = float(input("enter your second number"))
  if d != 0:
    print(x%d)
    startScreen()
  elif x != 0:
    print(d%x)
    startScreen()
  else:
    errorScreen()

  return


def factorial(w):
    c = 1
    w = int(input())
    for i in range(1, w + 1):
        c = c * i
        print(c)
        startScreen()
    return

def errorScreen():
  print("\n#######################################################")
  print("#                                                       #")
  print("#                                                       #")
  print("#                                                       #")
  print("                  error try again                       #")
  print("#                                                       #")
  print("#                                                       #")
  print("#                                                       #")
  print("#########################################################")
  return

startScreen()
```
